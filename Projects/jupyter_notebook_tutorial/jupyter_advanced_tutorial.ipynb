### Configuring Logging

```python
import logging
logging.error('Jupyter notebook')
```

```python
print('python')
logging.error('Jupyter')
logging.warning('notebook')
'print something'
```

```python
logger = logging.getLogger()
logger.setLevel(logging.DEBUG)

logging.info('This is some information')
logging.debug('This is a debug message')
```

```python
handler = logging.StreamHandler()
handler.setLevel(logging.DEBUG)

formater = logging.Formatter('%(levelname)s: %(message)s')
handler.setFormatter(formater)

logger.handlers = [handler]

logging.error('An error')
logging.warning('An warning')
logging.info('An info')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("tips")
```

```python
data.head()
```

```python
plt.scatter(data.total_bill, data.tip);
```

```python
sns.set(style="darkgrid")
plt.scatter(data.total_bill, data.tip);
```

```python
sns.scatterplot(x="total_bill", y="tip", data=data);
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=data);
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="size", style="smoker", data=data);
```

### Working with Databases

```python
%load_ext sql
```

```python
%sql sqlite://
```

```python
tips = sns.load_dataset("tips")
%sql PERSIST tips
```

```python
%%sql
SELECT * FROM tips
LIMIT 3
```

```python
result = %sql SELECT * FROM tips WHERE total_bill > (SELECT AVG(total_bill) FROM tips)
larger_bills = result.DataFrame()
larger_bills.head(3)
```
